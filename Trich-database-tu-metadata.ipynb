```python
import os
import json
import sqlite3

# Path to the directory containing your JSON files
json_directory = r"E:\University\AI_Challenge\AIC24\Data_BTC\media-info"
# SQLite database file
db_file = r"E:\University\AI_Challenge\AIC24\Data_BTC\video_metadata.db"

# Create a database connection
conn = sqlite3.connect(db_file)
cursor = conn.cursor()

# Create the Videos table
cursor.execute('''
CREATE TABLE IF NOT EXISTS Videos (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    video_id TEXT UNIQUE,
    length INTEGER,
    watch_url TEXT
)
''')

# Load data from JSON files
# Get a list of JSON files and sort them
json_files = [f for f in os.listdir(json_directory) if f.endswith('.json')]
json_files.sort()  # Sort the filenames

# Process each sorted JSON file
for filename in json_files:
    file_path = os.path.join(json_directory, filename)
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
        video_id = filename[:-5]  # Remove the .json extension

        # Insert data into the Videos table
        cursor.execute('''
        INSERT OR IGNORE INTO Videos (video_id, length, watch_url) 
        VALUES (?, ?, ?)
        ''', (
            video_id,
            data.get('length'),
            data.get('watch_url')
        ))

# Commit and close the connection
conn.commit()
conn.close()
print("Data has been loaded into the database in the correct order.")
```
